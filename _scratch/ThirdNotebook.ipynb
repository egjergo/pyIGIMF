```python
# Import necessary libraries
import IGIMF
import numpy as np
import pandas as pd
from scipy import optimize
import scipy.integrate as integr
```

```python
# Set up your input parameters (metallicity and downsizing galaxy stellar mass)
mass_metals: float = 1e7 # [Msun]
mass_gas: float = 1e9 # [Msun]
M_igal: float = 1e10 # [Msun]
```

```python
# Define an instance for a single delta t 
igimf = IGIMF.IGIMF(mass_metals, mass_gas, M_igal)#, downsizing_time, t, suppress_warnings=False)
```

```python
k_ecl, M_max, ECMF_func, ECMF_weighted_func = igimf.ECMF()
```

```python
resolution = int(100)
M_ecl_v = np.logspace(np.log10(igimf.M_ecl_min), np.log10(igimf.M_ecl_max), num=resolution)
ECMF_v = np.array([ECMF_func(M) for M in M_ecl_v])
ECMF_weighted_v = np.array([ECMF_weighted_func(M) for M in M_ecl_v])
```

```python
IGIMF.IMF_plot(M_ecl_v, ECMF_v, k_ecl, M_max, igimf.Mtot, r'ECMF [$M_{\odot}$]')
```

```python
IGIMF.IMF_plot(M_ecl_v, ECMF_weighted_v, k_ecl, M_max, igimf.Mtot, r'ECMF weighted [$M_{\odot}$]')
```
